```python
import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt

# Define the range for the x values
x = np.linspace(-4, 4, 1000)

# Calculate the PDF and CDF
pdf = norm.pdf(x)
cdf = norm.cdf(x)

# Plot the PDF and CDF
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF')
plt.title('Standard Normal Distribution - PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='orange')
plt.title('Standard Normal Distribution - CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()
```
