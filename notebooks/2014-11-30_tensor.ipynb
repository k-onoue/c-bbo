```python
import numpy as np

# Define a sample constraint function
def constraint_function(value):
    # Example constraint: Values must be >= 0, otherwise mark as -10
    return value if value >= 0 else -10

# Create a sample tensor (e.g., trial evaluations)
sample_tensor = np.array([
    [0.5, -0.2, 1.0],
    [1.5, -1.0, 0.0],
    [-0.3, 0.7, -0.1]
])

# Apply the constraint function element-wise using np.vectorize
tensor_constraint = np.vectorize(constraint_function)(sample_tensor)

# Display the original tensor and the constrained tensor
print("Original Tensor:")
print(sample_tensor)

print("\nConstrained Tensor:")
print(tensor_constraint)
```

```python
import torch

def build_constraint_warcraft() -> torch.Tensor:
    # Directions dictionary
    directions_dict = {
        "oo": torch.tensor([0, 0]),
        "ab": torch.tensor([1, 1]),
        "ac": torch.tensor([0, 2]),
        "ad": torch.tensor([1, 1]),
        "bc": torch.tensor([1, 1]),
        "bd": torch.tensor([2, 0]),
        "cd": torch.tensor([1, 1]),
    }

    directions_list = list(directions_dict.keys())

    # Map parameters
    map_shape = (2, 3)
    map_length = map_shape[0] * map_shape[1]
    ideal_gain = (map_shape[0] + map_shape[1] - 1) * 2

    # Initialize constraints as PyTorch tensors
    tensor_constraint_1 = torch.zeros((len(directions_list),) * map_length)
    tensor_constraint_2 = torch.zeros((len(directions_list),) * map_length)
    tensor_constraint_3 = torch.zeros((len(directions_list),) * map_length)

    # Constraint 1: (0, 0) != "oo", "ab"
    for direction in directions_list:
        if direction not in ["oo", "ab"]:
            tensor_constraint_1[..., directions_list.index(direction)] = 1

    # Constraint 2: (map_shape[0] - 1, map_shape[1] - 1) != "oo", "cd"
    for direction in directions_list:
        if direction not in ["oo", "cd"]:
            tensor_constraint_2[directions_list.index(direction), ...] = 1

    # Constraint 3: len[path] == map_shape[0] * map_shape[1]
    for index in torch.cartesian_prod(*[torch.arange(len(directions_list)) for _ in range(map_length)]):
        index_tuple = tuple(index.tolist())
        gain = sum(directions_dict[directions_list[idx]].sum().item() for idx in index_tuple)
        if gain == ideal_gain:
            tensor_constraint_3[index_tuple] = 1

    # Combine constraints with logical AND
    tensor_constraint = torch.logical_and(
        tensor_constraint_1, 
        torch.logical_and(tensor_constraint_2, tensor_constraint_3)
    )

    return tensor_constraint.float()


tensor_constraint = build_constraint_warcraft()
tensor_constraint
```
