```python
import os

import pandas as pd
import plotly.graph_objects as go
```

```python
date_dir_list = os.listdir("results_acqf_dist")
date_dir_list = sorted(date_dir_list)
date_dir_list
```

```python
target_csv_list = os.listdir("results_acqf_dist/2025-01-25_14-04-22")
target_csv_list = sorted(target_csv_list)
target_csv_list = [f for f in target_csv_list if f.endswith(".csv") and "loss" not in f]
target_csv_list
```

```python
target_date_and_filename_dict = {}

for date_dir in date_dir_list:
    target_csv_list = os.listdir(f"results_acqf_dist/{date_dir}")
    target_csv_list = sorted(target_csv_list)
    target_csv_list = [f for f in target_csv_list if f.endswith(".csv") and "loss" not in f]
    target_date_and_filename_dict[date_dir] = target_csv_list

target_date_and_filename_dict
```

```python
mid = int(len(target_date_and_filename_dict) / 2)

c_target_date_and_filename_dict = {key: target_date_and_filename_dict[key] for key in list(target_date_and_filename_dict.keys())[mid:]}
u_target_date_and_filename_dict = {key: target_date_and_filename_dict[key] for key in list(target_date_and_filename_dict.keys())[:mid]}
```

```python
task_dict = {
    "task1": "warcraft_map1",
    "task2": "warcraft_map2",
    "task3": "ackley_dim2",
}

split_data = {
    "task1": {},
    "task2": {},
    "task3": {},
}

# 振り分け
for key, files in c_target_date_and_filename_dict.items():
    for task_key, task_value in task_dict.items():
        if any(task_value in filename for filename in files):
            split_data[task_key][key] = files

split_data
```
